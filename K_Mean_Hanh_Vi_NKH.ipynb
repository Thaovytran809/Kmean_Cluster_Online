```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')

display(df)
df.describe()
```

```python
# Xử lý null
import missingno as msno

# Tính toán số lượng giá trị null trong mỗi cột
null_counts = df.isnull().sum()

# Tạo một DataFrame mới để hiển thị kết quả
null_df = pd.DataFrame({'Số lượng giá trị null': null_counts})

# Sắp xếp theo số lượng giá trị null giảm dần
null_df = null_df.sort_values(by='Số lượng giá trị null', ascending=False)

# Hiển thị kết quả
print(null_df)
```

```python
# Xử lý missing values cho Income
df.drop_duplicates()
df = df.dropna(subset=['Income'])
df = df.reset_index(drop=True)

# Kiểm tra lại số lượng giá trị null sau khi xoá
print(df.isnull().sum())

display(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Loại bỏ các cột không có ý nghĩa phân cụm
df.drop(columns=['Z_CostContact', 'Z_Revenue'], inplace=True)

# Tạo cột tuổi (Age)
df['Age'] = 2025 - df['Year_Birth']
df.drop(columns=['Year_Birth'], inplace=True)

# Tạo cột thời gian gắn bó (Customer_Since) tính bằng số ngày
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df['Customer_Since'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days
df.drop(columns=['Dt_Customer'], inplace=True)

# Gộp AcceptedCmp1 → AcceptedCmp5 thành AcceptedCmpTotal
accepted_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df['AcceptedCmpTotal'] = df[accepted_cols].sum(axis=1)
df.drop(columns=accepted_cols, inplace=True)

display(df)
df.describe()
```

```python
# số lượng outliers
def number_of_outliers(df):

    df = df.select_dtypes(exclude = 'object')

    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    return ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()

print(number_of_outliers(df))
```

```python
#remove outlier
def remove_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    print(f"{column_name} - Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    # Lưu số lượng giá trị bị loại bỏ
    initial_count = df.shape[0]
    df_filtered = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    removed_count = initial_count - df_filtered.shape[0]

    print(f"Removed outliers: {removed_count} from {initial_count} in {column_name}")

    return df_filtered

df = remove_outliers(df, 'NumWebPurchases')
df = remove_outliers(df, 'Age')
# giá trị của các biến phân loại thường đc định trc không có giá trị cực lớn hoặc cực nhỏ
```

```python
df["Marital_Status"] = df["Marital_Status"].replace(["Married","Together"],"Partner")
df["Marital_Status"] = df["Marital_Status"].replace(["Divorced","Widow","Alone","Absurd","YOLO"],"Single")
df["Education"] = df["Education"].replace(["Master","2n Cycle","PhD"],"PostGraduate")
df["Education"] = df["Education"].replace(["Graduation"],"Graduate")
df["Education"] = df["Education"].replace(["Basic"],"UnderGraduate")
```

```python
from sklearn.preprocessing import StandardScaler

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn import metrics #chỉ số đánh giá
```

```python
# Xác định số cụm tối ưu bằng phương pháp Elbow
inertias = []

for n_clusters in range(2, 10): #chạy thử số cụm từ 2 đến 10
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)

# Tạo biểu đồ Elbow để biểu thị số cụm tối ưu
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(2, 10), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = MiniBatchKMeans(n_init=10)
visualizer = KElbowVisualizer(model, k=(2,10), timings=False)
visualizer.fit(df_scaled)
visualizer.show()
```

Chọn K tối ưu là 3 ==> Phân ra 3 nhóm khách hàng

```python
# Train lại mô hình với số cụm tối ưu đã tìm đc
kmeans = MiniBatchKMeans(n_clusters=3, random_state=42, batch_size=64)
kmeans.fit(df_scaled)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Giảm chiều dữ liệu bằng PCA
pca = PCA(n_components=2)
data_2d = pca.fit_transform(df_scaled)
centroids_2d = pca.transform(kmeans.cluster_centers_)
```

```python
# Vẽ biểu đồ trực quan dữ liệu sau khi đã phân cụm
colors = ['red', 'blue', 'yellow']

plt.scatter(data_2d[:, 0], data_2d[:, 1], c=[colors[label] for label in kmeans.labels_]) # Vẽ điểm dlieu
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], c='black', marker='x', s=100, label='Centroids')  # Vẽ tâm cụm
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# Tính chỉ số Silhouette
sil_score = silhouette_score(df_scaled, df['Cluster'])
print(f"Chỉ số Silhouette: {sil_score:.4f}")
```

```python
# Thêm cột chứa nhãn cụm vào dữ liệu
df['Cluster'] = kmeans.labels_
display(df)
```
