```python
import pandas as pd

df = pd.read_csv('marketing_campaign.csv', sep='\t')

display(df)
df.describe()
```

```python
# Xá»­ lÃ½ null
import missingno as msno

# TÃ­nh toÃ¡n sá»‘ lÆ°á»£ng giÃ¡ trá»‹ null trong má»—i cá»™t
null_counts = df.isnull().sum()

# Táº¡o má»™t DataFrame má»›i Ä‘á»ƒ hiá»ƒn thá»‹ káº¿t quáº£
null_df = pd.DataFrame({'Sá»‘ lÆ°á»£ng giÃ¡ trá»‹ null': null_counts})

# Sáº¯p xáº¿p theo sá»‘ lÆ°á»£ng giÃ¡ trá»‹ null giáº£m dáº§n
null_df = null_df.sort_values(by='Sá»‘ lÆ°á»£ng giÃ¡ trá»‹ null', ascending=False)

# Hiá»ƒn thá»‹ káº¿t quáº£
print(null_df)
```

```python
# Xá»­ lÃ½ missing values cho Income
df.drop_duplicates()
df = df.dropna(subset=['Income'])
df = df.reset_index(drop=True)

# Kiá»ƒm tra láº¡i sá»‘ lÆ°á»£ng giÃ¡ trá»‹ null sau khi xoÃ¡
print(df.isnull().sum())

display(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1. Loáº¡i bá» cÃ¡c cá»™t khÃ´ng cÃ³ Ã½ nghÄ©a phÃ¢n cá»¥m
df.drop(columns=['Z_CostContact', 'Z_Revenue'], inplace=True)

# Táº¡o cá»™t tuá»•i (Age)
df['Age'] = 2025 - df['Year_Birth']
df.drop(columns=['Year_Birth'], inplace=True)

# Táº¡o cá»™t thá»i gian gáº¯n bÃ³ (Customer_Since) tÃ­nh báº±ng sá»‘ ngÃ y
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df['Customer_Since'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days
df.drop(columns=['Dt_Customer'], inplace=True)

# Gá»™p AcceptedCmp1 â†’ AcceptedCmp5 thÃ nh AcceptedCmpTotal
accepted_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df['AcceptedCmpTotal'] = df[accepted_cols].sum(axis=1)
df.drop(columns=accepted_cols, inplace=True)

display(df)
df.describe()
```

```python
# sá»‘ lÆ°á»£ng outliers
def number_of_outliers(df):

    df = df.select_dtypes(exclude = 'object')

    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1

    return ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()

print(number_of_outliers(df))
```

```python
#remove outlier
def remove_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    print(f"{column_name} - Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
    # LÆ°u sá»‘ lÆ°á»£ng giÃ¡ trá»‹ bá»‹ loáº¡i bá»
    initial_count = df.shape[0]
    df_filtered = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    removed_count = initial_count - df_filtered.shape[0]

    print(f"Removed outliers: {removed_count} from {initial_count} in {column_name}")

    return df_filtered

df = remove_outliers(df, 'NumWebPurchases')
df = remove_outliers(df, 'Age')
# giÃ¡ trá»‹ cá»§a cÃ¡c biáº¿n phÃ¢n loáº¡i thÆ°á»ng Ä‘c Ä‘á»‹nh trc khÃ´ng cÃ³ giÃ¡ trá»‹ cá»±c lá»›n hoáº·c cá»±c nhá»
```

```python
df["Marital_Status"] = df["Marital_Status"].replace(["Married","Together"],"Partner")
df["Marital_Status"] = df["Marital_Status"].replace(["Divorced","Widow","Alone","Absurd","YOLO"],"Single")
df["Education"] = df["Education"].replace(["Master","2n Cycle","PhD"],"PostGraduate")
df["Education"] = df["Education"].replace(["Graduation"],"Graduate")
df["Education"] = df["Education"].replace(["Basic"],"UnderGraduate")
```

```python
from sklearn.preprocessing import StandardScaler

# Chuáº©n hÃ³a dá»¯ liá»‡u
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn import metrics #chá»‰ sá»‘ Ä‘Ã¡nh giÃ¡
```

```python
# XÃ¡c Ä‘á»‹nh sá»‘ cá»¥m tá»‘i Æ°u báº±ng phÆ°Æ¡ng phÃ¡p Elbow
inertias = []

for n_clusters in range(2, 10): #cháº¡y thá»­ sá»‘ cá»¥m tá»« 2 Ä‘áº¿n 10
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)

# Táº¡o biá»ƒu Ä‘á»“ Elbow Ä‘á»ƒ biá»ƒu thá»‹ sá»‘ cá»¥m tá»‘i Æ°u
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(2, 10), inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = MiniBatchKMeans(n_init=10)
visualizer = KElbowVisualizer(model, k=(2,10), timings=False)
visualizer.fit(df_scaled)
visualizer.show()
```

Chá»n K tá»‘i Æ°u lÃ  3 ==> PhÃ¢n ra 3 nhÃ³m khÃ¡ch hÃ ng

```python
# Train láº¡i mÃ´ hÃ¬nh vá»›i sá»‘ cá»¥m tá»‘i Æ°u Ä‘Ã£ tÃ¬m Ä‘c
kmeans = MiniBatchKMeans(n_clusters=3, random_state=42, batch_size=64)
kmeans.fit(df_scaled)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Giáº£m chiá»u dá»¯ liá»‡u báº±ng PCA
pca = PCA(n_components=2)
data_2d = pca.fit_transform(df_scaled)
centroids_2d = pca.transform(kmeans.cluster_centers_)
```

```python
# Váº½ biá»ƒu Ä‘á»“ trá»±c quan dá»¯ liá»‡u sau khi Ä‘Ã£ phÃ¢n cá»¥m
colors = ['red', 'blue', 'yellow']

plt.scatter(data_2d[:, 0], data_2d[:, 1], c=[colors[label] for label in kmeans.labels_]) # Váº½ Ä‘iá»ƒm dlieu
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], c='black', marker='x', s=100, label='Centroids')  # Váº½ tÃ¢m cá»¥m
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
# TÃ­nh chá»‰ sá»‘ Silhouette
sil_score = silhouette_score(df_scaled, df['Cluster'])
print(f"Chá»‰ sá»‘ Silhouette: {sil_score:.4f}")
```

```python
# ThÃªm cá»™t chá»©a nhÃ£n cá»¥m vÃ o dá»¯ liá»‡u
df['Cluster'] = kmeans.labels_
display(df)
```

```python
# TÃ­nh trung bÃ¬nh cá»§a cÃ¡c Ä‘áº·c Ä‘iá»ƒm theo tá»«ng cá»¥m
cluster_means = df.groupby('Cluster').mean()
print(cluster_means)
```
